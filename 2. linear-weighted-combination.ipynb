# Linear Weighted Combination

> 여러 변수마다 가중치를 다르게 주어 정보를 혼합합니다. 3차원의 두 벡터를 가진 집합이 생성하는 부분공간은 2차원 평면입니다.

```python
inner_product = r"""
    \alpha v1 + \beta v2
"""

display(Math(inner_product))
```

```python
v1 = np.array([1, 2, 3])
v2 = np.array([2, -1, 1])

alpha = np.linspace(-2, 2, 10)
beta = np.linspace(-2, 2, 10)
X, Y = np.meshgrid(alpha, beta)
Z = v1[2] * X + v2[2] * Y

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(v1[0] * X + v2[0] * Y, v1[1] * X + v2[1] * Y, Z, alpha=0.5, color='cyan')

ax.quiver(0, 0, 0, v1[0], v1[1], v1[2], color='r', label="v1", linewidth=2)
ax.quiver(0, 0, 0, v2[0], v2[1], v2[2], color='b', label="v2", linewidth=2)

ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
ax.set_title("The subspace of the vector")

plt.legend()
plt.show()
```
